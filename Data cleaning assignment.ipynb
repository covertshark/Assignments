```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df.head()
```

```python
## 1.) values in the the FlightNumber column are missing.
df["FlightNumber"]
```

```python
df['FlightNumber']= df['FlightNumber'].interpolate().astype(int)
# df['FlightNumber']= df['FlightNumber']
df
```

```python
## 2.) The From_To column would be better as two separate columns.
temp_df= df
temp_df[['From', 'To']]= temp_df['From_To'].str.split("_",expand=True)
temp_df
```

```python
## 3.) Standardise the strings so that only the first letter is uppercase
temp_df['From']= temp_df['From'].str.capitalize()

temp_df['To']= temp_df['To'].str.capitalize()
temp_df
```

```python
## 4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions
df.drop(['From_To'], axis=1)
```

```python
temp_df['From_To']= temp_df[['From','To']].agg('_'.join, axis=1)
```

```python
temp_df
```

```python
df['From_To']= temp_df['From_To']
```

```python
df
```

```python
## 5.) RecentDelays column
delay= df['RecentDelays'].apply(pd.Series)
print(delay)
n=len(delay.columns)
delay.columns= ['delay {}'.format(n) for n in range(1, n+1)]
df= df.drop('RecentDelays', axis=1).join(delay)
```
