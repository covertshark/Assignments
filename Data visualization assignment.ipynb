```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import datasets
import plotly.express as px
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
df.head()
```

```python
x =df.iloc[:,0:4]
y= df.iloc[:,4]
```

```python
pca = PCA(n_components=3)
transform= pca.fit_transform(x)
new_df = pd.DataFrame(data = transform, columns = ['TD1', 'TD2', 'TD3'])
new_df.head()
```

```python
finaldf = pd.concat([new_df, df[['species']]], axis = 1)
```

```python
finaldf
```

```python
fig = px.scatter_3d(new_df, x='TD1', y='TD2', z='TD3', color=y)
fig.show()
```
